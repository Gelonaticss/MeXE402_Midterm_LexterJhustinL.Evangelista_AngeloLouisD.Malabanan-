### Step 1: Data Preprocessing

### Importing the dataset

```python
import pandas as pd
dataset = pd.read_csv('diabetes.csv')
```

```python
dataset.info
```

```python
dataset.head(10)
```

### Getting the Inputs and Outputs

```python
#[row,column]
X = dataset.iloc[:,:-1].values
y = dataset.iloc[:,-1].values
```

```python
X
```

```python
y
```

### Creating the Training Set and the Test Set

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size =0.2, random_state = 0)
```

```python
X_train
```

```python
X_test
```

```python
y_train
```

```python
y_test
```

### Feature Scalling

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
```

```python
X_train
```

### Building and Training the logistic regression model

### Building the model

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(random_state = 0)
```

### Training the model

```python
model.fit(X_train, y_train)
```

### Inference

Making the predictons of the data points in the test set

```python
y_pred = model.predict(sc.transform(X_train))
```

```python
y_pred
```

```python
y_test
```

Making the prediction of a single data point with:

1. Pregnancies = 6
2. Glucose = 148
3. Blood Pressure = 72
4. Skin Thickness = 35
5. Insulin = 0
6. BMI = 33.6
7. Diabetes Pedigree Function = 0.627
8. Age = 50


```python
model.predict(sc.transform([[6, 148, 72, 35, 0, 33.6, 0.627, 50]]))
```

### Evaluation Metrics

### Calculate accuracy

```python
from sklearn.metrics import accuracy_score
y_pred = model.predict(X_test)
accuracy_score(y_test, y_pred)
```

### Visualization

CONFUSION MATRIX

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, y_pred)
```

### Interpretation

### Model Classification Ability

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```
